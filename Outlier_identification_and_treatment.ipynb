```python
import numpy as np
import pandas as pd
```

```python
distribution = np.arange(1,100)
```

```python
print(distribution)
```

```python
outliers = [1000,155,199,-1,1000000000000000000]
```

```python
final_data = np.append(distribution,outliers)
df = pd.DataFrame(final_data)
```

```python
df.rename(columns={0:'data'},inplace=True)
```

## Now First lets see how to identify outliers

```python
# IQR Method

q1 = df['data'].quantile(0.25)
q3 = df['data'].quantile(0.75)

IQR = q3 - q1

lower_bound = q1 - (1.5*IQR)
upper_bound = q3 + (1.5*IQR)

print(lower_bound,upper_bound)
```

```python
outliers_ = (df[df['data'] < lower_bound]) | (df[df['data'] > upper_bound])
outliers_.head()
```

```python
df.iloc[outliers_.index]
```

## Now lets find the method to Treat this Outliers

```python
df1 = df.copy()
```

```python
# Method1 - To replace the outliers with median

median = df1['data'].median()
df1.loc[outliers_.index,'data'] = median
```

```python
df1.iloc[outliers_.index]
```

```python
# Method 2 - Winsorization(Capping the lower and upper bound)
'''
Winsorization ka matlab hai outliers ko replace karna threshold values se,
taaki data ka distribution preserved rahe aur extreme values ka impact kam ho jaye.
Isme outliers ko remove nahi kiya jata, balki unhe upper aur lower bounds ke nearest values se clip kar diya jata hai.
'''

df2 = df.copy()

df2 = np.clip(df2['data'], upper_bound, lower_bound)
df2.iloc[outliers_.index]
```

```python
# Metod 3 - Applying Log Transformation

df3 = df.copy()

df3['data'] = np.log(df3['data'])
df3.iloc[outliers_.index]
```
